This simulation aims to check if motoneurons are firing according to the correct recruitment order.

```python
import sys
sys.path.insert(0, '..')
# Allows plots to be zoomed in, etc
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
```

```python
spikeInstantMN = []
unitNumberMN = []
spikeInstantRC = []
unitNumberRC = []

filename = 'MNspk.dat'
f = open(filename, 'r')
lines = f.readlines()
for line in lines:
    spikeInstantMN.append(float(line.split()[0]))
    unitNumberMN.append(int(float(line.split()[1])))
f.close()

filename = 'INspk.dat'
f = open(filename, 'r')
lines = f.readlines()
for line in lines:
    spikeInstantRC.append(float(line.split()[0]))
    unitNumberRC.append(int(float(line.split()[1])))
f.close()
```

```python
FRS = []
FRFR = []
FRFF = []
S = [y for x, y in enumerate(spikeInstantMN) if unitNumberMN[x]==40]
FR = [y for x, y in enumerate(spikeInstantMN) if unitNumberMN[x]==120]
FF = [y for x, y in enumerate(spikeInstantMN) if unitNumberMN[x]==200]

for j in range(len(S)-1):
    FRS = np.append(FRS, [1000/(S[j+1]-S[j])])
del S[0]
for j in range(len(FR)-1):
    FRFR = np.append(FRFR, [1000/(S[j+1]-S[j])])
del FR[0]
for j in range(len(FF)-1):
    FRFF = np.append(FRFF, [1000/(S[j+1]-S[j])])
del FF[0]

plt.figure()
plt.plot(S, FRS, label='S')
plt.plot(FR, FRFR, label='FR')
plt.plot(FF, FRFF, label='FF')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(spikeInstantMN, unitNumberMN, '.')#, label = "Motoneuronios")
plt.title ('"final", Pconn with RCsd = 3.5/3, decaying weight, no dynamics')
plt.legend()
plt.xlabel('t (ms)')
plt.ylabel('# do Neuronio')
plt.show()

plt.figure()
plt.plot(spikeInstantRC, unitNumberRC, '.')#, label = "Motoneuronios")
plt.title ('"final", Pconn with RCsd = 3.5/3, decaying weight, no dynamics')
plt.legend()
plt.xlabel('t (ms)')
plt.ylabel('# do Neuronio')
plt.show()
```
